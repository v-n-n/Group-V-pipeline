```python
import numpy as np
import pandas as pd

# Create sample dataset
np.random.seed(42)
data = {
    'Student': [f'Student_{i}' for i in range(1, 51)],
    'Math_Score': np.random.randint(60, 100, 50),
    'Science_Score': np.random.randint(55, 98, 50),
    'English_Score': np.random.randint(50, 95, 50),
    'Hours_Studied': np.random.uniform(1, 8, 50).round(1)
}

df = pd.DataFrame(data)
print("Dataset created successfully!")
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
```

```python
# Calculate descriptive statistics
print("=== DESCRIPTIVE STATISTICS ===\n")

# Overall statistics
print("Overall Summary Statistics:")
print(df.describe())

# Subject averages
print("\n=== SUBJECT AVERAGES ===")
print(f"Math Average: {df['Math_Score'].mean():.2f}")
print(f"Science Average: {df['Science_Score'].mean():.2f}")
print(f"English Average: {df['English_Score'].mean():.2f}")

# Correlation analysis
print("\n=== CORRELATION MATRIX ===")
print(df[['Math_Score', 'Science_Score', 'English_Score', 'Hours_Studied']].corr())

# Additional insights
print("\n=== ADDITIONAL INSIGHTS ===")
print(f"Total Students: {len(df)}")
print(f"Average Hours Studied: {df['Hours_Studied'].mean():.2f}")
print(f"Max Math Score: {df['Math_Score'].max()}")
print(f"Min English Score: {df['English_Score'].min()}")
```
