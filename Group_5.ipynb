```python
import numpy as np
import pandas as pd

# Create sample dataset
np.random.seed(42)
data = {
    'Student': [f'Student_{i}' for i in range(1, 51)],
    'Math_Score': np.random.randint(60, 100, 50),
    'Science_Score': np.random.randint(55, 98, 50),
    'English_Score': np.random.randint(50, 95, 50),
    'Hours_Studied': np.random.uniform(1, 8, 50).round(1)
}

df = pd.DataFrame(data)
print("Dataset created successfully!")
print(f"Shape: {df.shape}")
print("\nFirst 5 rows:")
print(df.head())
```
